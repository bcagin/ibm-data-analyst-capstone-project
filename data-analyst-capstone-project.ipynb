```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m1_survey_data.csv"
```

```python
df=pd.read_csv(dataset_url)
```

```python
df.head()
```

```python
df["Age"].mean()
```

```python
df["Country"].nunique()
```

```python
df.columns
```

```python
df.LanguageWorkedWith.isnull().sum()
```

```python
df = df.dropna(subset=['LanguageWorkedWith'])
```

```python
all_languages_c = df['LanguageWorkedWith'].dropna().str.split(';')
```

```python
languages_series_c = all_languages_c.explode()
```

```python
language_counts_c = languages_series_c.value_counts()
```

```python
top10_languages_c = language_counts_c.head(10)
```

```python
plt.figure(figsize=(10,6))
top10_languages_c.plot(kind='bar')

plt.title("Top 10 Programming Languages (Current Year)")
plt.xlabel("Programming Language")
plt.ylabel("Number of Developers")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
